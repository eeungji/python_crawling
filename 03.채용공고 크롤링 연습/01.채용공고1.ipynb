```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.sgsilver.or.kr/board_employment01/view.php?tn=board_employment01&list_count_s=&G_state=Y&pm=5.1&part_part_s=&subject_s=&content_s=&name_s=&writedate_s=&keyfield_s=&search_word_s=&uid=206&gid=204&cpage=2&spage=1")
html = response.text
soup = BeautifulSoup(html, 'html.parser')
title = soup.select_one(".title.blue").text
file = soup.select_one(".files > a").attrs['href']
print(title, file)
```
